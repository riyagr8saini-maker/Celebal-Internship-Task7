```python
import pandas as pd

df = pd.read_csv('superstore.csv', encoding='windows-1252')
print("Data loaded successfully!")
```

```python
print("--- Data Exploration ---")

display(df.head())
display(df.tail())

print(f"\nDataset Shape (Rows, Columns): {df.shape}")

print("\nColumns in Dataset:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)
```

```python
print("--- Cleaning and Filtering ---")


print("Missing values before cleaning:\n", df.isnull().sum())

df_clean = df.dropna()

tech_df = df_clean[df_clean['Category'] == 'Technology']
tech_summary = tech_df[['Order ID', 'Category', 'Sales', 'Quantity']]

print(f"\nFiltered Technology Rows: {tech_df.shape[0]}")
display(tech_summary.head())
```

```python
print("--- Duplicates and Derived Columns ---")

df_clean = df_clean.drop_duplicates()
print(f"Row count after dropping duplicates: {df_clean.shape[0]}")

df_clean['Unit_Price'] = df_clean['Sales'] / df_clean['Quantity']

df_clean['Total_Amount'] = df_clean['Unit_Price'] * df_clean['Quantity']

display(df_clean[['Quantity', 'Unit_Price', 'Total_Amount']].head())
```

```python
print("--- Saving Data ---")

df_clean.to_csv('cleaned_superstore.csv', index=False)
print("File saved successfully as 'cleaned_superstore.csv'!")
```
